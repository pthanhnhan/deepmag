takeing frames out of video

```python
import cv2
import os
# Opens the Video file
cap= cv2.VideoCapture('/content/baby.mp4')
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
   
    cv2.imwrite((os.path.join('/content/data/vids/baby', str("%06d" % i)+'.png')),frame)
    i+=1

cap.release()
```
